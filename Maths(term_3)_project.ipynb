## Import Required Libraries

```python
import pandas as pd # Import the pandas library for data manipulation
import numpy as np  # Import the numpy library for numerical operations
```

## Load Dataset

```python
df = pd.read_csv(
    "air.csv.csv", # Specify the path to the CSV file
    sep=';',         # Define the separator used in the CSV file
    decimal=','      # Define the decimal separator used in the CSV file
)
```

## Dataset Information

## Missing Value Analysis

Missing values can affect analysis and therefore must be identified before preprocessing.

```python
df.isnull().sum() # Calculate and display the count of missing values for each column
```

```python
df.replace(-200, np.nan, inplace=True) # Replace all occurrences of -200 with NaN (Not a Number) to standardize missing values
```

```python
df.isnull().sum() # Re-calculate and display the count of missing values for each column after replacement
```

## Filling Missing Values

Numerical missing values are replaced using the median value of each column.
Median is preferred because it is less affected by outliers.

```python
num_cols = df.select_dtypes(include='number').columns # Select columns with numerical data types

df[num_cols] = df[num_cols].fillna( # Fill missing values in numerical columns
    df[num_cols].median()             # Use the median of each column for imputation
)
```

```python
df.isnull().sum() # Re-calculate and display the count of missing values for each column after filling
```

## Duplicate Value Analysis

Duplicate records may affect statistical analysis and should be identified.

```python
print(df.duplicated().sum()) # Print the total number of duplicate rows in the DataFrame
```

## Removing Duplicate Values

Duplicate records are removed to improve data quality and analysis accuracy.

```python
df = df.drop_duplicates() # Remove duplicate rows from the DataFrame
```

```python
print(df.duplicated().sum()) # Print the total number of duplicate rows after removal, expected to be 0
```
